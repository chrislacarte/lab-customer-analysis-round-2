# Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. _Hint_: If data from March does not exist, consider only January and February

```python
import time
import datetime 
file1_cat['month'] = pd.DatetimeIndex(file1_cat['effective to date']).month
file1_cat['month'].unique()
file1_cat[(file1_cat['month']==1) | (file1_cat['month']==2) | (file1_cat['month']==3)]


#or 

# file1_cat['month'] = file1_cat["effective to date"].dt.month_name()#
# file1_cat[(file1_cat["month"]=="January") | (file1_cat["month"]=="February") | (file1_cat["month"]=="March")]
```

```python
def clean(x):
    lst = []
    for col in file1.columns:
        lst.append(col.lower())
    file1.columns = lst

    #Deal with nan values
    file1['months since last claim'] = file1['months since last claim'].fillna(file1['months since last claim'].mean())
    file1['st'] = file1['st'].fillna('U')
    file1['response'] = file1['response'].fillna('No')
    file1['number of open complaints'] = file1['number of open complaints'].fillna(0)
    file1['vehicle class'] = file1['vehicle class'].fillna('U')
    file1['vehicle size'] =file1['vehicle size'].fillna('U')
    file1['vehicle type'] = file1['vehicle size'].fillna('U')
    
    #Chrnage coli=umn to datetime and extract month
    file1['effective to date'] = pd.to_datetime(file1['effective to date'], errors='coerce')
    file1['month'] = pd.DatetimeIndex(file1['effective to date']).month
    file1['month'].value_counts()
    file1['month'] = file1['month'].astype(int)
    filt = file1[(file1['month'].isin([1,2,3]))]
    
    return filt

```

```python
df3 = clean(file1)
```

# Show a plot of the response rate by the total claim amount.


We create a column where all 'yes' from response becomes 1, and 'No' becomes 0

```python
#resp_mean = df3['resp_rate'].mean()
#resp_mean
```

We calculate the range of "total claim amount" to find out whats the dimesion, so we can divide in bins>

```python
max_tcm = df3["total claim amount"].max()
min_tcm = df3['total claim amount'].min()

print(max_tcm)
print(min_tcm)
```

This func transforms a column of "continuous" vble into bins, with some params as inpouts. I could find the way of soft code the bin labels, so I hard coded them :)

```python
def categorize(col,min_val,max_val):
    
    bins = np.linspace(min_val, max_val,10)
    return pd.cut(col, bins, labels=[1,2,3,4,5,6,7,8,9])

df4 = categorize(df3['total claim amount'],0,3000)
print(df4)
```

Now we are cocatenating the df with "total claim amount" in bins to the original DF as a new column called 'claim_bins'

```python
df3['claim_bins'] = df4
```

Now we countplot the "claim_bins' with hue='response' => this will separate YES and NO responses by color ( visual response rate )

```python
sns.countplot(data=df3, x="claim_bins",hue= 'response')
```
